```python
def extract_placeholder_mapping(original_text, transformed_text, allowed_types):
    # STEP 1: ë¼ì¸ ë‹¨ìœ„ ë¶„ë¦¬
    orig_lines = original_text.splitlines()
    trans_lines = transformed_text.splitlines()
    # ì˜ˆì‹œ 
    # orig_lines[0]: ìƒë‹´ì‚¬: "ì•ˆë…•í•˜ì„¸ìš”, ê¹€ë¯¸ì˜ ê³ ê°ë‹˜. ë³´í—˜ ìƒë‹´ì„ ë„ì™€ë“œë¦´ í™ì„±ì² ì…ë‹ˆë‹¤. ìƒì¼ì´ 1985ë…„ 4ì›” 12ì¼ë¡œ ë“±ë¡ë˜ì–´ ìˆìŠµë‹ˆë‹¤. ë§ìœ¼ì‹ ê°€ìš”?"
    # trans_lines[0]: ìƒë‹´ì‚¬: "ì•ˆë…•í•˜ì„¸ìš”, [PERSON1] ê³ ê°ë‹˜. ë³´í—˜ ìƒë‹´ì„ ë„ì™€ë“œë¦´ [PERSON2]ì…ë‹ˆë‹¤. ìƒì¼ì´ [DATEOFBIRTH1]ë¡œ ë“±ë¡ë˜ì–´ ìˆìŠµë‹ˆë‹¤. ë§ìœ¼ì‹ ê°€ìš”?"

    # STEP 2: ê° ë¼ì¸ì—ì„œ placeholderì™€ ì¼ë°˜ í…ìŠ¤íŠ¸(literal)ë¥¼ êµ¬ë¶„
    allowed_pattern = re.compile(r'\[(' + '|'.join(allowed_types) + r')\d*\]')
    generic_pattern = re.compile(r'(\[[^]]+\])')

    mapping = {}
    # ì˜ˆì‹œ mapping['ê¹€ë¯¸ì˜'] = '[PERSON1]'

    n_lines = min(len(orig_lines), len(trans_lines))

    for idx in range(n_lines):
        orig_line = orig_lines[idx]
        trans_line = trans_lines[idx]

        parts = re.split(generic_pattern, trans_line)
        # ì˜ˆì‹œ 
        # parts = re.split(generic_pattern, trans_lines[0])
        # parts = [
        #     'ìƒë‹´ì‚¬: "ì•ˆë…•í•˜ì„¸ìš”, ', 
        #     '[PERSON1]', 
        #     ' ê³ ê°ë‹˜. ë³´í—˜ ìƒë‹´ì„ ë„ì™€ë“œë¦´ ', 
        #     '[PERSON2]', 
        #     'ì…ë‹ˆë‹¤. ìƒì¼ì´ ', 
        #     '[DATEOFBIRTH1]', 
        #     'ë¡œ ë“±ë¡ë˜ì–´ ìˆìŠµë‹ˆë‹¤. ë§ìœ¼ì‹ ê°€ìš”?"'
        # ]

        orig_pos = 0

        for i, part in enumerate(parts):
            if allowed_pattern.match(part):
                # placeholder ë°œê²¬
                # ë‹¤ìŒ literalì„ ì°¾ìŒ
                next_literal = parts[i + 1] if i + 1 < len(parts) else ''
                
                # ë‹¤ìŒ literalì´ ì¡´ì¬í•˜ë©´, ê·¸ literalê¹Œì§€ì˜ í…ìŠ¤íŠ¸ë¥¼ ì¶”ì¶œ
                if next_literal:
                    next_idx = orig_line.find(next_literal, orig_pos)
                    if next_idx != -1:
                        replaced_text = orig_line[orig_pos:next_idx]
                        orig_pos = next_idx
                    else:
                        # ë‹¤ìŒ literalì„ ëª» ì°¾ìœ¼ë©´ ëê¹Œì§€
                        replaced_text = orig_line[orig_pos:]
                        orig_pos = len(orig_line)
                else:
                    # ë‹¤ìŒ literalì´ ì—†ìœ¼ë©´ ë‚¨ì€ í…ìŠ¤íŠ¸ ì „ì²´
                    replaced_text = orig_line[orig_pos:]
                    orig_pos = len(orig_line)

                replaced_text = replaced_text.strip()
                if replaced_text:
                    mapping[replaced_text] = part

            else:
                # literalì¸ ê²½ìš°, ì›ë³¸ì—ì„œ ìœ„ì¹˜ ì—…ë°ì´íŠ¸
                found_idx = orig_line.find(part, orig_pos)
                if found_idx != -1:
                    orig_pos = found_idx + len(part)

    return mapping
```

```python
# OpenAIë¡œ ìƒì„±í•  ë•Œ ` ë¼ëŠ” íŠ¹ìˆ˜ë¬¸ìê°€ ìƒì„±ë˜ëŠ” ê²½ìš°ê°€ ìˆì–´ì„œ ì œê±° -> í‰ê°€í• ë•Œ ì˜¤ì‘ë™ì„ ì¼ìœ¼í‚µë‹ˆë‹¤. 
df["anonymized_data"] = df["anonymized_data"].map(lambda x: x.replace("`", ""))

# extract_placeholder_mappingë¥¼ applyì™€ í•¨ê»˜ ì‚¬ìš©í•´ì„œ mapping ì»´ëŸ¼ì„ ì¶”ê°€í•©ë‹ˆë‹¤. 
df["mapping"] = df.apply(lambda x: extract_placeholder_mapping(
    x["origin_data"], 
    x["anonymized_data"], 
    allowed_types=(
        "PERSON", "CONTACT", "ADDRESS", "ACCOUNT", "DATEOFBIRTH", 
        "EMAIL", "LOCATION", "KAKO_ID", "TWITTER_ID", "TELEGRAM_ID")), 
    axis=1)
df["mapping"] = df["mapping"].map(lambda x:str(x))
```

```python
# í•™ìŠµí•œ ëª¨ë¸ì„ ê²½ë¡œë¥¼ ì§€ì •í•©ë‹ˆë‹¤.
save_dir = "./model/model_0.0002_alpha-32_r-64"
peft_model_id = f"{save_dir}"

# PEFT ì–´ëŒ‘í„°ë¥¼ í†µí•´ ì‚¬ì „ í•™ìŠµëœ ëª¨ë¸ì„ ë¡œë“œí•©ë‹ˆë‹¤.
fine_tuned_model = AutoPeftModelForCausalLM.from_pretrained(
  peft_model_id,
  device_map="auto",
  torch_dtype=torch.float16
).to("cuda")

# í† í¬ë‚˜ì´ì € ë¡œë“œí•©ë‹ˆë‹¤.
tokenizer = AutoTokenizer.from_pretrained(peft_model_id)
tokenizer.padding_side = 'right'  
tokenizer.pad_token = tokenizer.eos_token
```

```python
import datasets 

# ë°ì´í„°ì…‹ì˜ ê° ìƒ˜í”Œì„ ì±—ë´‡ í•™ìŠµì— ë§ëŠ” í¬ë§·ìœ¼ë¡œ ë³€í™˜í•˜ëŠ” í•¨ìˆ˜ ì •ì˜
def get_chat_format(element):
    # ì‹œìŠ¤í…œ í”„ë¡¬í”„íŠ¸ë¥¼ ì •ì˜ (Assistantê°€ í•´ì•¼í•  ì‘ì—… ì„¤ëª…)
    system_prompt = "ë„ˆëŠ” ê°œì¸ì •ë³´ë¥¼ ë¹„ì‹ë³„í™”í•˜ëŠ” Assistantì•¼. ë„ˆëŠ” ì£¼ì–´ì§„ ë°ì´í„°ë¥¼ ë°”íƒ•ìœ¼ë¡œ ê°œì¸ì •ë³´ë¥¼ ë¹„ì‹ë³„í™”í•˜ëŠ” ì‘ì—…ì„ í•´ì•¼í•´."

    # ì±—ë´‡ ë©”ì‹œì§€ í¬ë§·ìœ¼ë¡œ ë°ì´í„° ë³€í™˜
    return {
        "messages": [
            {"role": "system", "content": system_prompt},                 # ì‹œìŠ¤í…œ ì§€ì‹œì‚¬í•­
            {"role": "user", "content": element["origin_data"]},          # ì›ë³¸ ë°ì´í„° (ì‚¬ìš©ì ì…ë ¥)
            {"role": "assistant", "content": element["anonymized_data"]}, # ë¹„ì‹ë³„í™”ëœ ë°ì´í„° (ëª¨ë¸ ì¶œë ¥ ì˜ˆì‹œ)
        ], 
        "label": element["mapping"]  # ì›ë³¸ê³¼ ë¹„ì‹ë³„í™”ëœ ë°ì´í„° ê°„ì˜ ë§¤í•‘(ì •ë‹µ ë¼ë²¨)
    }

# pandas ë°ì´í„°í”„ë ˆì„(df)ì„ Hugging Face Dataset í˜•íƒœë¡œ ë³€í™˜
dataset = datasets.Dataset.from_pandas(df)

# ì „ì²´ ë°ì´í„°ì…‹ì— get_chat_format í•¨ìˆ˜ë¥¼ ì ìš©í•˜ì—¬ ì±—ë´‡ í•™ìŠµ ë°ì´í„°ë¡œ ë³€í™˜
dataset = dataset.map(get_chat_format, remove_columns=dataset.features, batched=False)

# ë°ì´í„°ì˜ ìˆœì„œë¥¼ ë¬´ì‘ìœ„ë¡œ ì„ìŒ (ì‹œë“œê°’ì„ ì •í•´ ì¼ê´€ì„± ìœ ì§€) -> ë‹¤ì–‘í•œ ë°ì´í„°ê°€ ìˆê¸° ë•Œë¬¸ì— ì´ë¥¼ ì„ê¸° ìœ„í•´ 
dataset = dataset.shuffle(seed=42)

# ì „ì²´ ë°ì´í„°ì…‹ì„ í•™ìŠµìš©ê³¼ í‰ê°€ìš© ë°ì´í„°ë¡œ ë‚˜ëˆ” (í…ŒìŠ¤íŠ¸ì…‹ 10%, í•™ìŠµì…‹ 90%)
dataset = dataset.train_test_split(test_size=0.1, seed=42)
```

```python
# í…ŒìŠ¤íŠ¸ ë°ì´í„°ì…‹ì„ í•˜ë‚˜ì”© ìˆœíšŒí•˜ë©´ì„œ í‰ê°€ë¥¼ ì§„í–‰
for conv_data in dataset["test"]:
    # tokenizerì˜ ì±— í…œí”Œë¦¿ì„ ì´ìš©í•˜ì—¬ ì…ë ¥ ë©”ì‹œì§€(ì‹œìŠ¤í…œ+ì‚¬ìš©ì)ë¥¼ ìì—°ìŠ¤ëŸ¬ìš´ ì±— í˜•ì‹ìœ¼ë¡œ ë³€í™˜
    # tokenize=Falseë¡œ ì„¤ì •í•˜ì—¬ í…ìŠ¤íŠ¸ í˜•íƒœë¡œ ë°˜í™˜
    # add_generation_prompt=Trueë¡œ ì„¤ì •í•˜ì—¬ Assistantê°€ ë‹µë³€í•  ì¤€ë¹„ë¥¼ í•¨
    input_data = tokenizer.apply_chat_template(
        conv_data["messages"][:2], tokenize=False, add_generation_prompt=True
    )

    # ë³€í™˜ëœ í…ìŠ¤íŠ¸ ë°ì´í„°ë¥¼ ëª¨ë¸ ì…ë ¥ìœ¼ë¡œ ì‚¬ìš©í•˜ê¸° ìœ„í•´ í† í°í™”í•˜ê³  GPU(CUDA)ì— ë¡œë”©
    inputs = tokenizer(input_data, return_tensors="pt").to("cuda")

    # fine-tuned ëª¨ë¸ì„ ì‚¬ìš©í•˜ì—¬ ë¹„ì‹ë³„í™”ëœ ê²°ê³¼ë¥¼ ìƒì„±
    result = fine_tuned_model.generate(
        **inputs,                      # í† í°í™”ëœ ì…ë ¥ ì „ë‹¬
        max_new_tokens=512,            # ìƒì„±í•  ìµœëŒ€ í† í° ê°œìˆ˜ ì„¤ì •
        temperature=0.1,               # ë‚®ì€ temperatureë¡œ ì˜ˆì¸¡ ê²°ê³¼ì˜ ëœë¤ì„±ì„ ë‚®ì¶¤ (ê²°ê³¼ê°€ ì¼ê´€ë¨)
        pad_token_id=tokenizer.eos_token_id  # ë¬¸ì¥ ëì„ ë‚˜íƒ€ë‚´ëŠ” í† í° ì„¤ì • (íŒ¨ë”© ë°©ì§€)
    )

    # ìƒì„±ëœ ê²°ê³¼ì—ì„œ ì…ë ¥ ë¶€ë¶„ì€ ì œì™¸í•˜ê³  ìƒˆë¡œ ìƒì„±ëœ ê²°ê³¼(ëª¨ë¸ì˜ ì‘ë‹µ)ë§Œ ì¶”ì¶œí•˜ì—¬ í…ìŠ¤íŠ¸ë¡œ ë³€í™˜
    output = tokenizer.decode(
        result[0][len(inputs.input_ids[0]):],  # ì…ë ¥ ë¶€ë¶„ í† í°ì„ ì œì™¸í•œ ìƒì„±ëœ ë¶€ë¶„ë§Œ ì¶”ì¶œ
        skip_special_tokens=True               # íŠ¹ìˆ˜ í† í°ì„ ì œì™¸í•˜ê³  í…ìŠ¤íŠ¸ ë³€í™˜
    )

    # í•œ ë²ˆë§Œ ì‹¤í–‰í•˜ê³  ë°˜ë³µë¬¸ ì¢…ë£Œ (í…ŒìŠ¤íŠ¸ë¥¼ ìœ„í•œ ì˜ˆì‹œ)
    break
```

```python
# ì‹¤ì œ ì •ë‹µ ë§¤í•‘(Ground Truth)ê³¼ ëª¨ë¸ì´ ì˜ˆì¸¡í•œ ë§¤í•‘ ê²°ê³¼ë¥¼ ë¹„êµí•˜ì—¬ ì •í™•ë„ë¥¼ í‰ê°€í•˜ëŠ” í•¨ìˆ˜
def compare_mappings(ground_truth, prediction, consider_partial_match=True):
    # 1. ë‘ ë§¤í•‘ì˜ í‚¤(ê°œì¸ì •ë³´ í•­ëª©)ë¥¼ ë¹„êµ
    gt_keys = set(ground_truth.keys())  # ì •ë‹µì˜ í‚¤ë“¤
    pred_keys = set(prediction.keys())  # ëª¨ë¸ ì˜ˆì¸¡ì˜ í‚¤ë“¤

    common_keys = gt_keys.intersection(pred_keys)  # ì •ë‹µê³¼ ì˜ˆì¸¡ ëª¨ë‘ì— ì¡´ì¬í•˜ëŠ” í‚¤
    only_in_gt = gt_keys - pred_keys               # ì •ë‹µì—ë§Œ ì¡´ì¬í•˜ëŠ” í‚¤ (ëˆ„ë½ëœ í•­ëª©)
    only_in_pred = pred_keys - gt_keys             # ì˜ˆì¸¡ì—ë§Œ ì¡´ì¬í•˜ëŠ” í‚¤ (ì¶”ê°€ëœ í•­ëª©)

    # 2. ì™„ì „ ì¼ì¹˜í•˜ëŠ” í•­ëª©ì„ ê³„ì‚° (í‚¤ê°€ ê°™ê³  ê°’ë„ ì •í™•íˆ ê°™ì•„ì•¼ í•¨)
    correct_values = sum(1 for k in common_keys if ground_truth[k] == prediction[k])

    # 3. ë¶€ë¶„ ì¼ì¹˜ í•­ëª© ì°¾ê¸° (ê°’ì´ ì •í™•íˆ ì¼ì¹˜í•˜ì§€ë§Œ í‚¤ê°€ ì•½ê°„ ë‹¤ë¥¸ ê²½ìš°)
    partial_matches = []  # ë¶€ë¶„ ì¼ì¹˜ í•­ëª©ì˜ ìŒ (ì •ë‹µí‚¤, ì˜ˆì¸¡í‚¤)
    partial_correct = 0   # ë¶€ë¶„ ì¼ì¹˜ ê°œìˆ˜ ì¹´ìš´íŠ¸

    if consider_partial_match:
        # ì •ë‹µì—ë§Œ ìˆëŠ” í‚¤ë¥¼ ê¸°ì¤€ìœ¼ë¡œ ë¶€ë¶„ì ìœ¼ë¡œ ìœ ì‚¬í•œ ì˜ˆì¸¡ í‚¤ê°€ ìˆëŠ”ì§€ í™•ì¸
        for gt_key in list(only_in_gt):
            for pred_key in list(only_in_pred):
                # í‚¤ê°€ ì„œë¡œ í¬í•¨ê´€ê³„(ë¶€ë¶„ ë¬¸ìì—´ ê´€ê³„)ì´ë©´ ë¶€ë¶„ ì¼ì¹˜ë¡œ íŒë‹¨
                if (gt_key in pred_key or pred_key in gt_key):
                    if ground_truth[gt_key] == prediction[pred_key]:
                        partial_matches.append((gt_key, pred_key))
                        partial_correct += 1
                        only_in_gt.remove(gt_key)     # ë¶€ë¶„ ì¼ì¹˜ í™•ì¸ëœ í‚¤ëŠ” ëª©ë¡ì—ì„œ ì œê±°
                        only_in_pred.remove(pred_key) # ë¶€ë¶„ ì¼ì¹˜ í™•ì¸ëœ í‚¤ëŠ” ëª©ë¡ì—ì„œ ì œê±°
                        break

    # 4. í‰ê°€ ê²°ê³¼ ì¶œë ¥ (ìƒì„¸í•œ ë¶„ì„)
    print(f"ì™„ì „ ì¼ì¹˜ í‚¤ ìˆ˜: {len(common_keys)}/{len(gt_keys)} ({len(common_keys)/len(gt_keys)*100:.2f}%)")

    if consider_partial_match:
        print(f"ë¶€ë¶„ ì¼ì¹˜ í‚¤ ìŒ ìˆ˜: {len(partial_matches)}/{len(gt_keys)} ({len(partial_matches)/len(gt_keys)*100:.2f}%)")
        print(f"ë¶€ë¶„ ì¼ì¹˜ í‚¤ ìŒ: {partial_matches}")

    print(f"GTì—ë§Œ ìˆëŠ” í‚¤: {only_in_gt}")    # ëˆ„ë½ëœ í‚¤ ëª©ë¡
    print(f"ì˜ˆì¸¡ì—ë§Œ ìˆëŠ” í‚¤: {only_in_pred}") # ì¶”ê°€ëœ í‚¤ ëª©ë¡

    # 5. ì •í™•ë„ ê³„ì‚° (ì™„ì „ì¼ì¹˜ + ë¶€ë¶„ì¼ì¹˜ë¥¼ í¬í•¨í•˜ì—¬ ê³„ì‚°)
    total_correct = correct_values + partial_correct
    accuracy = total_correct / len(gt_keys) if gt_keys else 0
    print(f"ì „ì²´ ì •í™•ë„(ë¶€ë¶„ ì¼ì¹˜ í¬í•¨): {accuracy:.4f} ({total_correct}/{len(gt_keys)})")

    # ì™„ì „ ì¼ì¹˜ë§Œ ê³ ë ¤í•œ ì—„ê²©í•œ ì •í™•ë„ ê³„ì‚°
    strict_accuracy = correct_values / len(gt_keys) if gt_keys else 0
    print(f"ì—„ê²©í•œ ì •í™•ë„(ì™„ì „ ì¼ì¹˜ë§Œ): {strict_accuracy:.4f} ({correct_values}/{len(gt_keys)})")

    # ìµœì¢… ê²°ê³¼ë¥¼ dictionaryë¡œ ë°˜í™˜
    return {
        "accuracy": accuracy,
        "strict_accuracy": strict_accuracy,
        "common_keys": common_keys,
        "correct_values": correct_values,
        "partial_matches": partial_matches,
        "partial_correct": partial_correct,
        "only_in_gt": only_in_gt,
        "only_in_pred": only_in_pred
    }

# ì‹¤ì œ ë°ì´í„°ì—ì„œ ê°€ì ¸ì˜¨ ì •ë‹µ ë¼ë²¨(json í˜•íƒœ)ì„ dictionaryë¡œ ë³€í™˜
gt_mapping = json_repair.loads(conv_data["label"])

# ëª¨ë¸ì´ ì˜ˆì¸¡í•œ ë§¤í•‘ ê²°ê³¼ ì‚¬ìš©
pred_mapping = mapping_result

# ë‘ ë§¤í•‘ ê²°ê³¼ë¥¼ ë¹„êµí•˜ì—¬ ì •í™•ë„ í‰ê°€ ì‹¤í–‰
accuracy = compare_mappings(gt_mapping, pred_mapping)
```

```python
def compare_simple_mappings(ground_truth, prediction, sample_idx, consider_partial_match=True):
    eval_result = compare_mappings(ground_truth, prediction, consider_partial_match)
    eval_result["sample_idx"] = sample_idx
    return eval_result

def aggregate_results_simple(results):
    """
    ì—¬ëŸ¬ ê°œì˜ ìƒ˜í”Œì— ëŒ€í•œ ê°œë³„ í‰ê°€ ê²°ê³¼ë¥¼ í•˜ë‚˜ë¡œ ì¢…í•©í•˜ì—¬ ìš”ì•½í•˜ëŠ” í•¨ìˆ˜ì…ë‹ˆë‹¤.
    ê°„ê²°í•œ í˜•ì‹ìœ¼ë¡œ ì£¼ìš” ì§€í‘œë¥¼ ì§‘ê³„í•˜ì—¬ ë°˜í™˜í•©ë‹ˆë‹¤.
    """
    total_samples = len(results)  # ì „ì²´ í‰ê°€ ìƒ˜í”Œ ìˆ˜

    # ìƒ˜í”Œë³„ ì •í™•ë„ì™€ ì—„ê²©í•œ ì •í™•ë„ì˜ í‰ê·  ê³„ì‚°
    total_accuracy = sum(r["evaluation"]["accuracy"] for r in results) / total_samples
    strict_accuracy = sum(r["evaluation"]["strict_accuracy"] for r in results) / total_samples

    # ë¶€ë¶„ ì¼ì¹˜í•œ ì¼€ì´ìŠ¤ ìˆ˜ì§‘ (ìƒ˜í”Œ ì¸ë±ìŠ¤ í¬í•¨)
    all_partial_matches = []
    for r in results:
        for gt_key, pred_key in r["evaluation"]["partial_matches"]:
            all_partial_matches.append((r["sample_idx"], gt_key, pred_key))

    # ê°€ì¥ í”íˆ ë°œìƒí•œ ë¶€ë¶„ ì¼ì¹˜ íŒ¨í„´(ìƒ˜í”Œ ë²ˆí˜¸ì™€ í•¨ê»˜)ì„ ì§‘ê³„
    partial_match_patterns = {}
    for idx, gt_key, pred_key in all_partial_matches:
        pattern = f"ìƒ˜í”Œ {idx}: {gt_key} -> {pred_key}"
        partial_match_patterns[pattern] = partial_match_patterns.get(pattern, 0) + 1

    # ê°€ì¥ í”íˆ ë°œìƒí•œ ì˜¤ë¥˜(ì¶”ê°€/ëˆ„ë½) íŒ¨í„´ì„ ìƒ˜í”Œ ë²ˆí˜¸ì™€ í•¨ê»˜ ì§‘ê³„
    error_patterns = {}
    for r in results:
        # ì •ë‹µì—ë§Œ ìˆëŠ” í‚¤ (ëˆ„ë½)
        for gt_key in r["evaluation"]["only_in_gt"]:
            pattern = f"ìƒ˜í”Œ {r['sample_idx']}: ëˆ„ë½ : {gt_key}"
            error_patterns[pattern] = error_patterns.get(pattern, 0) + 1
        # ì˜ˆì¸¡ì—ë§Œ ìˆëŠ” í‚¤ (ì¶”ê°€)
        for pred_key in r["evaluation"]["only_in_pred"]:
            pattern = f"ìƒ˜í”Œ {r['sample_idx']}: ì¶”ê°€ : {pred_key}"
            error_patterns[pattern] = error_patterns.get(pattern, 0) + 1

    # ì§‘ê³„ëœ ê²°ê³¼ë¥¼ ë°˜í™˜
    return {
        "total_samples": total_samples,
        "average_accuracy": total_accuracy,
        "average_strict_accuracy": strict_accuracy,
        "common_partial_matches": sorted(partial_match_patterns.items(), key=lambda x: x[1], reverse=True),
        "common_errors": sorted(error_patterns.items(), key=lambda x: x[1], reverse=True)
    }

def print_evaluation_summary_comprehensive(eval_results):
    """
    í‰ê°€ ê²°ê³¼ ìš”ì•½ì„ ì¶œë ¥í•˜ëŠ” í•¨ìˆ˜ (ì¢…í•©ì ì¸ í‰ê°€ ì§€í‘œ í¬í•¨)
    """
    overall = eval_results["overall_results"]

    # ì „ì²´ ë°ì´í„°ì…‹ì— ëŒ€í•œ ì •í™•ë„ ë° ì¶”ê°€ ì§€í‘œ ê³„ì‚°ì„ ìœ„í•œ ì´ˆê¸°í™”
    total_gt_keys = 0              # ì´ ì •ë‹µ í‚¤ ê°œìˆ˜
    total_pred_keys = 0            # ì´ ì˜ˆì¸¡ í‚¤ ê°œìˆ˜
    total_correct_keys = 0         # ì™„ì „ ì¼ì¹˜í•œ í‚¤ ê°œìˆ˜
    total_partial_matches = 0      # ë¶€ë¶„ ì¼ì¹˜í•œ í‚¤ ê°œìˆ˜

    # ê°œì²´ ìœ í˜•ë³„ í†µê³„ ë°ì´í„°ë¥¼ ë‹´ì„ ë”•ì…”ë„ˆë¦¬
    entity_type_stats = {}

    # ê° ìƒ˜í”Œë³„ í‰ê°€ ê²°ê³¼ë¥¼ ìˆœíšŒí•˜ë©° í†µê³„ê°’ ê³„ì‚°
    for r in eval_results["sample_results"]:
        gt_mapping = r["ground_truth_mapping"]
        pred_mapping = r["predicted_mapping"]

        total_gt_keys += len(gt_mapping)
        total_pred_keys += len(pred_mapping)
        total_correct_keys += r["evaluation"]["correct_values"]
        total_partial_matches += r["evaluation"]["partial_correct"]

        # ì •ë‹µ ë°ì´í„°ì—ì„œ ê°œì²´ ìœ í˜•ë³„ í†µê³„ ìˆ˜ì§‘
        for key, value in gt_mapping.items():
            # ê°œì²´ ìœ í˜• ì¶”ì¶œ (ì˜ˆì‹œ: [PERSON1]ì—ì„œ PERSON)
            entity_type = re.match(r'\[([A-Z_]+)', value)
            if entity_type:
                entity_type = entity_type.group(1)
                if entity_type not in entity_type_stats:
                    entity_type_stats[entity_type] = {"gt": 0, "correct": 0, "partial": 0, "pred": 0}

                entity_type_stats[entity_type]["gt"] += 1

                # ì™„ì „ ì¼ì¹˜ ì—¬ë¶€ í™•ì¸
                if key in pred_mapping and pred_mapping[key] == value:
                    entity_type_stats[entity_type]["correct"] += 1

        # ì˜ˆì¸¡ ë°ì´í„°ì—ì„œ ê°œì²´ ìœ í˜•ë³„ í†µê³„ ìˆ˜ì§‘
        for key, value in pred_mapping.items():
            entity_type = re.match(r'\[([A-Z_]+)', value)
            if entity_type:
                entity_type = entity_type.group(1)
                if entity_type not in entity_type_stats:
                    entity_type_stats[entity_type] = {"gt": 0, "correct": 0, "partial": 0, "pred": 0}

                entity_type_stats[entity_type]["pred"] += 1

        # ë¶€ë¶„ ì¼ì¹˜ëœ ê°œì²´ ìœ í˜•ë³„ í†µê³„ ìˆ˜ì§‘
        for gt_key, pred_key in r["evaluation"]["partial_matches"]:
            value = gt_mapping[gt_key]
            entity_type = re.match(r'\[([A-Z_]+)', value)
            if entity_type:
                entity_type = entity_type.group(1)
                if entity_type in entity_type_stats:
                    entity_type_stats[entity_type]["partial"] += 1

    # ì „ì²´ í‰ê°€ ì§€í‘œ ê³„ì‚° (ì •í™•ë„, ì •ë°€ë„, ì¬í˜„ìœ¨, F1 ì ìˆ˜)
    accuracy = (total_correct_keys + total_partial_matches) / total_gt_keys if total_gt_keys > 0 else 0
    strict_accuracy = total_correct_keys / total_gt_keys if total_gt_keys > 0 else 0

    precision = total_correct_keys / total_pred_keys if total_pred_keys > 0 else 0
    recall = total_correct_keys / total_gt_keys if total_gt_keys > 0 else 0
    f1 = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0

    # ë¶€ë¶„ ì¼ì¹˜ë¥¼ í¬í•¨í•œ ìœ ì—°í•œ í‰ê°€ ì§€í‘œ ê³„ì‚°
    precision_with_partial = (total_correct_keys + total_partial_matches) / total_pred_keys if total_pred_keys > 0 else 0
    recall_with_partial = (total_correct_keys + total_partial_matches) / total_gt_keys if total_gt_keys > 0 else 0
    f1_with_partial = 2 * precision_with_partial * recall_with_partial / (precision_with_partial + recall_with_partial) if (precision_with_partial + recall_with_partial) > 0 else 0

    # ì¢…í•© í‰ê°€ ê²°ê³¼ ì¶œë ¥
    print("=" * 70)
    print(f"ì´ í‰ê°€ ìƒ˜í”Œ ìˆ˜: {overall['total_samples']}")

    # ì „ì²´ ë°ì´í„°ì…‹ í‰ê°€ ì§€í‘œ ì¶œë ¥
    print("\nì „ì²´ ë°ì´í„°ì…‹ í‰ê°€ ì§€í‘œ:")
    print(f"ì´ ì •ë‹µ í‚¤ ìˆ˜: {total_gt_keys}")
    print(f"ì´ ì˜ˆì¸¡ í‚¤ ìˆ˜: {total_pred_keys}")
    print(f"ì™„ì „ ì¼ì¹˜ í‚¤ ìˆ˜: {total_correct_keys} ({total_correct_keys/total_gt_keys*100:.2f}%)")
    print(f"ë¶€ë¶„ ì¼ì¹˜ í‚¤ ìˆ˜: {total_partial_matches} ({total_partial_matches/total_gt_keys*100:.2f}%)")

    # ì—„ê²©í•œ í‰ê°€ ì§€í‘œ ì¶œë ¥
    print("\nì—„ê²©í•œ í‰ê°€ (ì™„ì „ ì¼ì¹˜ë§Œ):")
    print(f"ì •í™•ë„(Accuracy): {strict_accuracy:.4f}")
    print(f"ì •ë°€ë„(Precision): {precision:.4f}")
    print(f"ì¬í˜„ìœ¨(Recall): {recall:.4f}")
    print(f"F1 ì ìˆ˜: {f1:.4f}")

    # ìœ ì—°í•œ í‰ê°€ ì§€í‘œ ì¶œë ¥
    print("\nìœ ì—°í•œ í‰ê°€ (ë¶€ë¶„ ì¼ì¹˜ í¬í•¨):")
    print(f"ì •í™•ë„(Accuracy): {accuracy:.4f}")
    print(f"ì •ë°€ë„(Precision): {precision_with_partial:.4f}")
    print(f"ì¬í˜„ìœ¨(Recall): {recall_with_partial:.4f}")
    print(f"F1 ì ìˆ˜: {f1_with_partial:.4f}")
```

```python
def evaluate_anonymization_model_simple(model, tokenizer, test_dataset, num_samples=None, temperature=0.1, max_new_tokens=512):
    """
    ëª¨ë¸ì˜ ë¹„ì‹ë³„í™” ì„±ëŠ¥ì„ í‰ê°€í•˜ëŠ” í•¨ìˆ˜ (ê°„ê²°í•œ ì¶œë ¥ ë²„ì „)
    """
    # ê²½ê³  ë©”ì‹œì§€ ì–µì œ (pad_token ê´€ë ¨)
    import warnings
    warnings.filterwarnings("ignore", message="Setting `pad_token_id` to")

    # í‰ê°€í•  ìƒ˜í”Œ ê°œìˆ˜ ì§€ì • (Noneì´ë©´ ì „ì²´ ë°ì´í„°ì…‹ ì‚¬ìš©)
    if num_samples is None:
        samples = test_dataset
    else:
        samples = test_dataset.select(range(min(num_samples, len(test_dataset))))

    results = []  # í‰ê°€ ê²°ê³¼ë¥¼ ì €ì¥í•  ë¦¬ìŠ¤íŠ¸

    # ë°ì´í„°ì…‹ì„ ìˆœíšŒí•˜ë©° í‰ê°€
    for idx, conv_data in enumerate(samples):
        print("------------------------------------")
        print(f"ìƒ˜í”Œ {idx+1}/{len(samples)} í‰ê°€ ì¤‘...")

        # ì…ë ¥ ë©”ì‹œì§€ë¥¼ ì±— í¬ë§·ìœ¼ë¡œ ë³€í™˜í•˜ì—¬ ì¤€ë¹„ (ì‹œìŠ¤í…œ í”„ë¡¬í”„íŠ¸ì™€ ì‚¬ìš©ì ì…ë ¥ë§Œ ì‚¬ìš©)
        input_data = tokenizer.apply_chat_template(
            conv_data["messages"][:2],
            tokenize=False,
            add_generation_prompt=True
        )

        # ì…ë ¥ ë°ì´í„°ë¥¼ í† í°í™”í•˜ì—¬ ëª¨ë¸ì— ì „ë‹¬
        inputs = tokenizer(input_data, return_tensors="pt").to(model.device)

        # ëª¨ë¸ì´ ì˜ˆì¸¡ì„ ìˆ˜í–‰í•˜ì—¬ ê²°ê³¼ ìƒì„±
        result = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            do_sample=True,
            temperature=temperature,
            pad_token_id=tokenizer.eos_token_id
        )

        # ëª¨ë¸ì˜ ìƒì„± ê²°ê³¼ë¥¼ í…ìŠ¤íŠ¸ë¡œ ë³€í™˜ (ì…ë ¥ ë¶€ë¶„ì€ ì œì™¸)
        output = tokenizer.decode(result[0][len(inputs.input_ids[0]):], skip_special_tokens=True)

        # ì›ë³¸ ì…ë ¥ í…ìŠ¤íŠ¸ ì¶”ì¶œ
        raw_input = conv_data["messages"][1]["content"]

        # ì›ë³¸ê³¼ ìƒì„± ê²°ê³¼ë¥¼ ë¹„êµí•˜ì—¬ placeholder ë§¤í•‘ ì¶”ì¶œ
        mapping_result = extract_placeholder_mapping(
            raw_input.replace("`", ""),
            output.replace("`", ""),
            allowed_types=(
                "PERSON", "CONTACT", "ADDRESS", "ACCOUNT", "DATEOFBIRTH",
                "EMAIL", "LOCATION", "KAKO_ID", "TWITTER_ID", "TELEGRAM_ID"
            )
        )

        # ì •ë‹µ ë§¤í•‘ ë°ì´í„° ë¡œë“œ
        gt_mapping = json_repair.loads(conv_data["label"])

        # ëª¨ë¸ ì˜ˆì¸¡ ê²°ê³¼ì™€ ì •ë‹µì„ ë¹„êµí•˜ì—¬ í‰ê°€ ìˆ˜í–‰
        evaluation = compare_simple_mappings(gt_mapping, mapping_result, sample_idx=idx)

        # ê°œë³„ ìƒ˜í”Œì˜ í‰ê°€ ê²°ê³¼ ì €ì¥
        results.append({
            "sample_idx": idx,
            "raw_input": raw_input,
            "model_output": output,
            "ground_truth_mapping": gt_mapping,
            "predicted_mapping": mapping_result,
            "evaluation": evaluation
        })

    # ëª¨ë“  ìƒ˜í”Œ í‰ê°€ ê²°ê³¼ë¥¼ ì¢…í•©í•˜ì—¬ ìš”ì•½
    overall_results = aggregate_results_simple(results)

    # ê°œë³„ ìƒ˜í”Œ í‰ê°€ ê²°ê³¼ì™€ ì¢…í•© í‰ê°€ ê²°ê³¼ ë°˜í™˜
    return {
        "sample_results": results,
        "overall_results": overall_results
    }

# ì‚¬ìš© ì˜ˆì‹œ
# ì „ì²´ í…ŒìŠ¤íŠ¸ ë°ì´í„°ì…‹ì„ ì‚¬ìš©í•˜ì—¬ í‰ê°€ ìˆ˜í–‰
eval_results = evaluate_anonymization_model_simple(fine_tuned_model, tokenizer, dataset["test"], num_samples=len(dataset["test"]))
```
